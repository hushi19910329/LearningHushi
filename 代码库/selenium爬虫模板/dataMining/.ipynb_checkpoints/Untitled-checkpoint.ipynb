```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
%matplotlib inline
```

```python
# .\mongod.exe --dbpath=H:\py7\scrapyxx\guomeiSele\db --port=27011
# 连接mongodb
def get_collection(ip='localhost',
                   port='27011',
                   db_name='hello',
                   collection_name='hello'):
    connection = f'mongodb://{ip}:{port}/'
    client = MongoClient(connection)
    db = client[db_name]
    collection = db[collection_name]
    return collection
```

```python
collection = get_collection()
```

```python
collection.find_one().keys()
```

```python
data = list(collection.find())
```

```python
data[0]
```

```python
df = pd.DataFrame(data)
# df = df.drop_duplicates('commentsSpots')
df.shape
```

```python
df = df[df['ItemName'].apply(lambda x:1 if '笔记本电脑' in x else 0)==1]
# df = df[df['品牌'].isnull()==False]
```

```python
def extrat_price(x):
    x = re.findall('\d+\.\d+',x)
    if len(x)>0:
        return x[0]
    else:
        return None
```

```python
df['ItemPrice'] = df['ItemPrice'].apply(extrat_price).astype(float)
df['ItemPrice'].plot.hist()
```

```python
tmp = df[df['ItemPrice']>10000]
```

```python
def extract_brands(x):
    brands = ['华硕','宏碁','神舟','戴尔','惠普','联想','外星人']
    for i in brands:
        if i in x:
            return i
```

```python
tmp['ItemName'] = tmp['ItemName'].apply(extract_brands)
tmp['ItemName'].value_counts().plot.bar()
```

```python
df['ItemName'] = df['ItemName'].apply(extract_brands)
```

```python
df.columns
```

```python
tmp = df.commentsSpots.apply(lambda x:x.split(' ')).values.tolist()
all_words = []
for line in tmp:
    all_words.extend(line)
all_words = [i for i in all_words if len(i)>0]
all_words
```

```python
labels, counts = np.unique(all_words, return_counts=True)
```

```python
from pyecharts import  WordCloud

help(WordCloud)
```

```python
wc = WordCloud()
wc.add('',labels, counts)
wc
```
